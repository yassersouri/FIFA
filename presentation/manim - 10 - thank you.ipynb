```python
%%manim -v WARNING Slides10ThankYou

class Slides10ThankYou(Scene):
    def construct(self):
        title = Tex(
            r"Thank You",
            font_size=70,
            color=BLUE
        )
        authors = Tex(
            r"Paper: arxiv.org/abs/2108.03894\\",
            r"~\\",
            r"Code: github.com/yassersouri/FIFA",
            font_size=40,
            tex_environment="flushleft"
        )
        
        texts = VGroup(title, authors).arrange(DOWN, buff=2, center=True).shift(UP)
        
        self.add(texts)
```
