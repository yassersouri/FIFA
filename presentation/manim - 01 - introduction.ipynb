```python
import numpy as np
from manim import *
config.media_width = "60%"
```

```python
import cv2
```

```python
bonn = cv2.imread("bonn_logo.png")[:, :, ::-1]
tme = cv2.imread("tme_logo.png")[:, :, ::-1]
```

```python
%%manim -v WARNING -qm Slides01Intro

class Slides01Intro(Scene):
    def construct(self):
        title = Tex(
            r"FIFA: Fast Inference Approximation\\ for Action Segmentation",
            font_size=70,
            color=BLUE
        )
        authors = Tex(
            r"\underline{Yaser Souri}, Yazan Abu Farha, \\Fabien Despinoy, Gianpiero Fransesca, Juergen Gall",
            font_size=40,
            tex_environment="flushleft"
        )
        
        texts = VGroup(title, authors).arrange(DOWN, buff=2, center=True).shift(UP)
        
        bonn_logo = ImageMobject(bonn, scale_to_resolution=1500).to_edge(DR)
        tme_logo = ImageMobject(tme, scale_to_resolution=1500).to_edge(DL)

        self.add(title)
        self.add(authors, bonn_logo, tme_logo)
```
